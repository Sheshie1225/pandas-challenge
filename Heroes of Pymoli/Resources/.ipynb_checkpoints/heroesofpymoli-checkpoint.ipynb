```python
import pandas as pd
import numpy as np
```

```python
file_to_load = "Resources/pymoli.csv"

pymoli = pd.read_csv("pymoli.csv")
pymoli.head()
```

```python
#Player Count
total_players = pymoli.SN.nunique()
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

```python
#Purchasing Analysis - Total
unique_items = len(pymoli["Item ID"].value_counts())
average_price = pymoli["Price"].mean()
purchases = pymoli["Purchase ID"].count()
total_revenue = pymoli["Price"].sum()

summarydf = pd.DataFrame({"Number of Unique Items":[unique_items],
                          "Average Price":[average_price],
                          "Number of Purchase":[purchases],
                          "Total Revenue":[total_revenue]
                         })
#formatting
summarydf.style.format({"Average Price":"${:.2f}", "Total Revenue": "${:,.2f}"})
                   
```

```python
#Gender Demographics
gender_df = pymoli[["SN", "Gender"]]
gender_df.shape

gender_df = gender_df.drop_duplicates()
gender_df.shape
```

```python
gender_count = gender_df["Gender"].value_counts()
percent_player = gender_count / total_players
```

```python
demographics = pd.DataFrame({"Total Number":gender_count,"Percentage of Players":percent_player})
demographics.style.format({"Percentage of Players":"{:.2%}"})
```

```python
#Purchasing Analysis - Gender

gender = pymoli.groupby("Gender")

purchase_count = gender["Purchase ID"].count()
purchase_price = gender["Price"].mean()
value = gender["Price"].sum()
purchase_per_person = value / gender_count

analysis = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":purchase_price, "Total Purchase Value":value, "Avg Total Purchase per Person":purchase_per_person})
analysis.style.format({"Average Purchase Price":"${:.2f}", "Total Purchase Value":"${:,.2f}", "Avg Total Purchase per Person":"${:.2f}"})
```

```python
#Age Demographics

agegroup_data = pymoli[["SN", "Age"]]
agegroup_data = agegroup_data.drop_duplicates()
agegroup_data.shape

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
agegroup_data["Age Groups"] = pd.cut(agegroup_data["Age"], bins, labels=bin_names)
agegroup_data.head(10)

agegroup = agegroup_data.groupby("Age Groups")
total_count = agegroup["Age"].count()
percentage = total_count / total_players

age_demo = pd.DataFrame({"Total Count":total_count, "Percentage of Players":percentage})
age_demo.style.format({"Percentage of Players":"{:.2%}"})
```

```python
#Purchasing Analysis - Age

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pymoli["Age Group"] = pd.cut(pymoli["Age"], bins, labels=bin_names)
age_group = pymoli.groupby("Age Group")
age_group.head()


purchase_count = age_group["Purchase ID"].count()
price = age_group["Price"].mean()
value = age_group["Price"].sum()
total_purchase = value / total_count

age_analysis = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":price, "Total Purchase Value":value, "Avg Total Purchase per Person":total_purchase})
age_analysis.style.format({"Average Purchase Price":"${:,.2f}", "Total Purchase Value":"${:,.2f}", "Avg Total Purchase per Person":"${:,.2f}"})
```

```python
#Top Spenders

spenders_df = pymoli.groupby(["SN"])

purchase_count = spenders_df["Purchase ID"].count()
price = spenders_df["Price"].mean()
value = spenders_df["Price"].sum()


top_spenders_df = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":price, "Total Purchase Value":value})

top_spenders_df = top_spenders_df.sort_values(["Total Purchase Value"], ascending=False)

top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]].head()
                            
```

```python
#Most Popular Items

top_items_df = pymoli.groupby(["Item ID", "Item Name"])

top_purchase_count = top_items_df["Purchase ID"].count()

top_items_total = top_items_df["Price"].sum()

top_item_price = top_items_total / top_purchase_count


top_items = pd.DataFrame({"Purchase Count":top_purchase_count, "Item Price": top_item_price, "Total Purchase Value":top_items_total})

item_summ = top_items.sort_values(["Purchase Count"], ascending=False)

item_summ["Item Price"] = item_summ["Item Price"].map("${:,.2f}".format)
item_summ["Total Purchase Value"] = item_summ["Total Purchase Value"].map("${:,.2f}".format)

item_summ.head()
```

```python
#Most Profitable Items

profitable_items = top_items.sort_values(["Total Purchase Value"], ascending = False)

profitable_items.head()
```
